```python
# 1.How do you calculate the mean, median, and mode of a dataset

import statistics

# Example dataset
data = [1, 2, 3, 3, 4, 5]

# Calculate Mean
mean = sum(data) / len(data)

# Calculate Median
median = statistics.median(data)

# Calculate Mode
try:
    mode = statistics.mode(data)
except statistics.StatisticsError:
    mode = "No unique mode"

# Output results
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
```

```python
# 5.Write a Python function to calculate the range of a dataset

def calculate_range(data):
    """
    Calculate the range of a dataset.

    Parameters:
        data (list): A list of numerical values.

    Returns:
        float: The range of the dataset.
    """
    if not data:
        return "Dataset is empty"

    # Calculate range
    data_range = max(data) - min(data)
    return data_range

# Example usage
dataset = [10, 20, 30, 40, 50]
result = calculate_range(dataset)
print(f"Range of the dataset: {result}")
```

```python
# 9. Write a Python script to calculate covariance between two datasets

import numpy as np

def calculate_covariance(dataset1, dataset2):
    """
    Calculate the covariance between two datasets.

    Parameters:
        dataset1 (list or array): The first dataset.
        dataset2 (list or array): The second dataset.

    Returns:
        float: The covariance value.
    """
    if len(dataset1) != len(dataset2):
        raise ValueError("Datasets must have the same length.")

    # Calculate the covariance
    covariance_matrix = np.cov(dataset1, dataset2, bias=True)  # bias=True calculates population covariance
    covariance = covariance_matrix[0, 1]  # Extract the covariance value

    return covariance

# Example usage
dataset1 = [10, 20, 30, 40, 50]
dataset2 = [15, 25, 35, 45, 55]

covariance = calculate_covariance(dataset1, dataset2)
print(f"Covariance between the datasets: {covariance:.2f}")
```

```python
# 10.Write a Python script to calculate the correlation coefficient between two datasets

import numpy as np

def calculate_correlation_coefficient(dataset1, dataset2):
    """
    Calculate the correlation coefficient between two datasets.

    Parameters:
        dataset1 (list or array): The first dataset.
        dataset2 (list or array): The second dataset.

    Returns:
        float: The correlation coefficient (r).
    """
    if len(dataset1) != len(dataset2):
        raise ValueError("Datasets must have the same length.")

    # Calculate the correlation coefficient
    correlation_matrix = np.corrcoef(dataset1, dataset2)
    correlation_coefficient = correlation_matrix[0, 1]  # Extract the coefficient

    return correlation_coefficient

# Example usage
dataset1 = [10, 20, 30, 40, 50]
dataset2 = [15, 25, 35, 45, 55]

correlation = calculate_correlation_coefficient(dataset1, dataset2)
print(f"Correlation Coefficient: {correlation:.2f}")
```

```python
#17.Calculate the interquartile range (IQR) of a dataset

import numpy as np

def calculate_iqr(data):
    """
    Calculate the Interquartile Range (IQR) of a dataset.

    Parameters:
        data (list or array): A list or array of numerical values.

    Returns:
        float: The IQR of the dataset.
    """
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)

    # Calculate IQR
    iqr = Q3 - Q1
    return iqr

# Example usage
dataset = [1, 3, 5, 7, 9, 10, 12, 14, 18, 20]
iqr = calculate_iqr(dataset)
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
#18. Implement Z-score normalization and explain its significance

import numpy as np

def z_score_normalization(data):
    """
    Apply Z-score normalization to a dataset.

    Parameters:
        data (list or array): A list or array of numerical values.

    Returns:
        array: Z-score normalized data.
    """
    mean = np.mean(data)
    std_dev = np.std(data)

    # Apply Z-score normalization
    normalized_data = (data - mean) / std_dev
    return normalized_data

# Example usage
dataset = [10, 20, 30, 40, 50]
normalized_data = z_score_normalization(np.array(dataset))

print("Original Data:", dataset)
print("Z-score Normalized Data:", normalized_data)
```

```python
#19.Compare two datasets using their standard deviations

import numpy as np

def compare_standard_deviation(data1, data2):
    """
    Compare the standard deviations of two datasets.

    Parameters:
        data1 (list or array): First dataset.
        data2 (list or array): Second dataset.

    Returns:
        tuple: Standard deviations of both datasets.
    """
    # Calculate standard deviation for each dataset
    std_dev1 = np.std(data1)
    std_dev2 = np.std(data2)

    # Compare
    if std_dev1 > std_dev2:
        result = "Dataset 1 has higher variability."
    elif std_dev1 < std_dev2:
        result = "Dataset 2 has higher variability."
    else:
        result = "Both datasets have equal variability."

    return std_dev1, std_dev2, result

# Example usage
dataset1 = [10, 20, 30, 40, 50]
dataset2 = [15, 25, 35, 45, 90]

std_dev1, std_dev2, result = compare_standard_deviation(dataset1, dataset2)

print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
print(f"Standard Deviation of Dataset 2: {std_dev2:.2}")
```

```python
#20. Write a Python program to visualize covariance using a heatmap

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
data = {
    "Age": [22, 35, 29, 40, 28],
    "Income": [30000, 45000, 37000, 52000, 40000],
    "Score": [85, 90, 75, 88, 92]
}

# Convert the dataset into a NumPy array for covariance calculation
data_array = np.array([data["Age"], data["Income"], data["Score"]])

# Calculate the covariance matrix
cov_matrix = np.cov(data_array)

# Heatmap visualization
plt.figure(figsize=(8, 6))
sns.heatmap(cov_matrix, annot=True, fmt=".2f", cmap="coolwarm", xticklabels=list(data.keys()), yticklabels=list(data.keys()))

# Add labels and title
plt.title("Covariance Heatmap")
plt.xlabel("Variables")
plt.ylabel("Variables")
plt.show()
```

```python
#22. Generate a dataset and implement both variance and standard deviation computations

import numpy as np

# Step 1: Generate a dataset
np.random.seed(42)  # For reproducibility
dataset = np.random.randint(10, 100, size=20)  # Random integers between 10 and 100

# Step 2: Compute variance
variance = np.var(dataset)  # Population variance
sample_variance = np.var(dataset, ddof=1)  # Sample variance (ddof=1)

# Step 3: Compute standard deviation
std_dev = np.std(dataset)  # Population standard deviation
sample_std_dev = np.std(dataset, ddof=1)  # Sample standard deviation (ddof=1)

# Output the results
print("Dataset:", dataset)
print(f"Population Variance: {variance:.2f}")
print(f"Sample Variance: {sample_variance:.2f}")
print(f"Population Standard Deviation: {std_dev:.2f}")
print(f"Sample Standard Deviation: {sample_std_dev:.2f}")
```

```python
#23.Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Step 1: Generate a dataset
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Normal distribution (mean=50, std=10)

# Add some outliers to show skewness and kurtosis effects
data = np.append(data, [10, 90, 100, 110])

# Step 2: Calculate skewness and kurtosis
skewness = skew(data)
kurt = kurtosis(data, fisher=True)  # Fisher=True gives excess kurtosis (normal dist = 0)

# Step 3: Visualize skewness and kurtosis
plt.figure(figsize=(12, 6))

# Plot the histogram
sns.histplot(data, bins=30, kde=True, color="skyblue", edgecolor="black")
plt.title(f"Skewness: {skewness:.2f}, Kurtosis: {kurt:.2f}")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.7)

# Show the plot
plt.show()
```

```python
#24.Implement the Pearson and Spearman correlation coefficients for a dataset

import numpy as np
from scipy.stats import spearmanr, pearsonr

# Example datasets
x = [1, 2, 3, 4, 5]  # Dataset 1
y = [10, 20, 30, 40, 50]  # Dataset 2 (perfect positive linear relationship)
z = [50, 40, 30, 20, 10]  # Dataset 3 (perfect negative relationship)

# Compute Pearson correlation coefficient
pearson_correlation_xy = pearsonr(x, y)[0]  # Pearson for x and y
pearson_correlation_xz = pearsonr(x, z)[0]  # Pearson for x and z

# Compute Spearman correlation coefficient
spearman_correlation_xy = spearmanr(x, y)[0]  # Spearman for x and y
spearman_correlation_xz = spearmanr(x, z)[0]  # Spearman for x and z

# Output results
print("Pearson Correlation Coefficient:")
print(f"x and y: {pearson_correlation_xy:.2f}")
print(f"x and z: {pearson_correlation_xz:.2f}")

print("\nSpearman Correlation Coefficient:")
print(f"x and y: {spearman_correlation_xy:.2f}")
print(f"x and z: {spearman_correlation_xz:.2f}")
```
